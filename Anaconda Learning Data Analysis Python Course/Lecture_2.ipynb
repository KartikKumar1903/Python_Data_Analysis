# If statement

```python
if <condotion>:
    <code>
elif <condition>:
    <code>
...
else:
    <code>
```

```python
age = 14

if age == 18:
    print("You're an adult!")
elif age >= 13:
    print("You're a teenager")
else:
    print("You're a kid")
```

# For Loop

```python
for <variable> in <iterable>:
    <code>
```

```python
countries = ["United States", "India", "China", "Brazil"]
```

```python
countries
```

```python
for country in countries:
    print(country)
```

```python
for i, country in enumerate(countries):
    print(f"{i}, {country}")
```

```python
my_data = {'name' : 'Frank', 'age' : 26}
my_data
```

```python
for key, value in my_data.items():
    print(f"{key} : {value}")
```

# Function

```python
def function(<params>):
    <code>
    return <data>
```

```python
def sum_values(a, b):
    x = a + b
    return x
```

```python
sum_values(1, 3)
```
