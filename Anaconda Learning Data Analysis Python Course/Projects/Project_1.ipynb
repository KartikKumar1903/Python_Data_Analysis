## Read .csv from Multiple URLs with Pandas

**Link : root + season + league**

```python
# Link structure

"https://www.football-data.co.uk/mmz4281/"+"2526"+"/"+"E0"+'.csv'
```

```python
# Create a root variable

root = "https://www.football-data.co.uk/mmz4281/"
```

### Multiple Leagues

- https://www.football-data.co.uk/mmz4281/2526/E0.csv
- https://www.football-data.co.uk/mmz4281/2526/E1.csv
- https://www.football-data.co.uk/mmz4281/2526/E2.csv
- https://www.football-data.co.uk/mmz4281/2526/E3.csv
- https://www.football-data.co.uk/mmz4281/2526/EC.csv


```python
# Creating a list of leagues

leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

# Looping through leagues, read multiple csv and append it into a list

for league in leagues:
    df = pd.read_csv(root + '2526' + '/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
```

```python
# Length of frames

len(frames)
```

```python
# Show elements

frames[0]
```

## Multiple Seasons

```python
# Creating list of leagues
leagues = ['E0', 'E2', 'EC']
frames = []

# Loop through range of seasons and insert new column
for league in leagues:
    for season in range(20, 26):
        df = pd.read_csv(root + str(season)+str(season+1) + '/' + league + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df)
        
```

```python
len(frames)
```

## Organize all data in dictionary

```python
# Create a dictionary with original name as leauge as key

dict_countries = {
    'Spansish La Liga' : 'SP1',
    'Spanish Segunda Division' : 'SP2',
    'German Bundesliga' : 'D1',
    'English Premier League' : 'E0',
    'English League' : 'E2'
}
```

```python
# Get a dictionary element

dict_countries['English League']
```

```python
dict_historical_data = {}

# Creating a list of leagues


# Looping through key elements
for league in dict_countries:
    frames = []
    for season in range(20, 26):
        df = pd.read_csv(root + str(season)+str(season+1)+'/'+ dict_countries[league] + '.csv', encoding="unicode_escape")
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
