# Pandas Attributes, Methods & Functions

## Attributes
---
A value associated with an object (referenced by name using dotted expressions)
```
df.columns # columns in an attribute
```

## Function
---
A group of related statements that performs a specific task.

Python built-in functions:

- max()
- min()
- len()

## Methods
---
A function which is defined inside a class body

``` 
df.head()
```

````python
import pandas as pd
````

````python
# Reading the csv file

df_exams = pd.read_csv('StudentsPerformance.csv')
````

````python
# Showing the dataframe

df_exams
````

# Attributes

````python
# Getting access to the shape attribute

df_exams.shape
````

````python
# Getting access to the index attribute

df_exams.index
````

````python
# Getting access to the column attribute

df_exams.columns
````

````python
# Data types of each column

df_exams.dtypes
````

# Methods

````python
# Showing the first 5 rows

df_exams.head()
````

````python
# Showing the info of the dataframe

df_exams.info()
````

````python
# Describing basic  statistics of the dataframe

df_exams.describe()
````

# Functions

````python
# Obtaining the length of the dataframe (number of rows)

len(df_exams)
````

````python
# Obtaining the highest index of the dataframe

max(df_exams.index)
````

````python
# Obtaining the lowest index of the dataframe

min(df_exams.index)
````

````python
# Obtaining the data type

type(df_exams)
````

````python
# Rounding the values of the dataset

round(df_exams, 2)
````
