```python
import pandas as pd
import numpy as np
```

```python
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# Math Operations

## Operations in columns

```python
# Select a column and calculate total sum

df_exams['math score'].sum()
```

```python
# count, mean, std, max and min

df_exams['math score'].count()
df_exams['math score'].mean()
df_exams['math score'].std()  # Standard deviation
df_exams['math score'].max()
df_exams['math score'].min()
```

```python
# Easier calculation with .describe()

df_exams.describe()
```

## Operations in Rows

```python
# Calculating the sum in row

df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
# Calculating the average score and assigning the result to a 
# new column

df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'])/3
```

```python
# Showing the dataframe

df_exams.round(2)
```

## Value Counts

```python
# Counting gender elements

# len function
len(df_exams['gender'])

# .count() method
df_exams['gender'].count()
```

```python
# Counting gender elements by category

df_exams['gender'].value_counts()
```

```python
# Return the relative frequency (divide all values by 
# the sum of values)

df_exams['gender'].value_counts(normalize = True)
```

```python
# Counting 'parental level of education' element by category

df_exams['parental level of education'].value_counts()
```

```python
# Return the relative frequency and round to 2 decimals

df_exams['parental level of education'].value_counts(normalize = True).round(2)
```

## Sort a Dataframe

```python
# Sort by one column

df_exams.sort_values(by = 'math score')
```

```python
# Sort descending by one column

df_exams.sort_values('math score', ascending=False)
```

```python
# Sort descending by multiple columns

# df_exams.sort_values(['math score', 'reading score'], 
                    # ascending = False) # works in same manner
df_exams.sort_values(by = ['math score', 'reading score'], 
                    ascending = False)
```

```python
# Sort descending bu multiple columns and update dataframe

df_exams.sort_values(['math score', 'reading score'], 
                     ascending = False,
                     inplace = True)
```

```python
df_exams
```

```python
# Sort descending with a key function
# We'll use the 'key' argument to ignore case of text
df_exams.sort_values('race/ethnicity', ascending=True, 
                     key=lambda col:col.str.lower())
```

## Create Index

```python
# Creating non repetitive values for the index

new_index = np.arange(0, 1000)
```

```python
# Shuffling the indexes

import random
random.shuffle(new_index)
```

```python
# Create a new column with new index

df_exams['new_index'] = new_index
df_exams
```

## Set Index

```python
# Set new_index column as index

df_exams.set_index('new_index', inplace=True)
```

## Sort Index

```python
# Sort dataframe by new index

df_exams.sort_index(ascending=True, inplace=True)
```

# Rename Column

```python
# Rename column and overwrite values of the dataframe

df_exams = df_exams.rename(columns={'gender' : 'Gender'})
```

```python
# Rename multiple columns and update the dataframe with 
# inplace arguement

df_exams.rename(columns={'math score' : 'MS',
                         'reading score' : 'RS',
                         'writing score' : 'WS'}, inplace=True)
```

```python
# Show dataframe

df_exams
```

## Rename Index

```python
# Rename index 0, 1 and 2 and update the dataframe

df_exams.rename(index={0:'A',
                       1:'B',
                       2:'C'}, inplace=True)
```
